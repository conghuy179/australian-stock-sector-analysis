```python
import pandas as pd

# Load the combined dataset from CSV
combined_df = pd.read_csv(r'C:\Users\congh\Downloads\DA Project\Project for Resume\ASX_Stock_Analysis_Project\data\processed\combined_stock_data.csv')

# Convert 'Date' column to datetime format
combined_df['Date'] = pd.to_datetime(combined_df['Date'])

# Show first 3 rows to confirm it's working
combined_df.head(3)
```

```python
# Filter data for the desired date range (e.g., 2010‚Äì2020)
filtered_df = combined_df[(combined_df['Date'] >= '2010-01-01') & (combined_df['Date'] <= '2020-12-31')]
print(filtered_df['Date'].min(), 'to', filtered_df['Date'].max())
print("Rows after filtering:", filtered_df.shape[0])
filtered_df.head()

```

```python
#Drop rows with missing key data
cleaned_df = filtered_df.dropna(subset=['Open', 'High', 'Low', 'Close', 'Adj Close', 'Volume'])

print("Rows after dropping missing values:",cleaned_df.shape[0])
```

```python
# Remove rows where Volume = 0
cleaned_df = cleaned_df[cleaned_df['Volume'] > 0]

print("Rows after removing Volume = 0:", cleaned_df.shape[0])
```

```python
# Round the adj close column for readibility and 0.01% accurate
cleaned_df['Adj Close'] = cleaned_df['Adj Close'].round(4)
print(cleaned_df['Adj Close'].head(10))
```

```python
#calculate stock-level performance
import numpy as np

# Get the first and last price per stock
stock_returns = cleaned_df.sort_values('Date').groupby('Stock').agg(
    start_price=('Adj Close', 'first'),
    end_price=('Adj Close', 'last'),
    start_date=('Date', 'first'),
    end_date=('Date', 'last'),
    Sector=('Sector', 'first')
)
# Fix: bring Stock back as a column
stock_returns = stock_returns.reset_index()

# Calculate number of years
stock_returns['Years'] = (stock_returns['end_date'] - stock_returns['start_date']).dt.days / 365.25

# Compound Annual Growth Rate formula
stock_returns['CAGR'] = (stock_returns['end_price'] / stock_returns['start_price']) ** (1 / stock_returns['Years']) - 1

# Drop any rows with bad values 
stock_returns = stock_returns.dropna(subset=['CAGR'])

# Group by sector
sector_performance = stock_returns.groupby('Sector')['CAGR'].agg(['mean', 'std', 'count']).sort_values(by='mean', ascending=False)

print(sector_performance)
```
